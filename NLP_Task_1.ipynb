# Task 1: Conceptual Understanding

1. What is the difference between "Love" and "love" in NLP?

To a computer, "Love" (with a capital L) and "love" (with a lowercase l) are treated as two completely different words because they have different character encodings (ASCII/Unicode values). If you do not convert all text to lowercase during preprocessing, the NLP model will learn them as separate tokens. This artificially increases the size of your vocabulary and dilutes the learned meaning of the word since its occurrences are split across two different variations. 

2. What happens if stopwords are not removed?

Stopwords are common words like "the", "is", "in", "and", and "to". If they are not removed:
Noise is added: They appear very frequently but carry almost no actual meaning about the topic of the text.Important words are overshadowed: In algorithms that rely on word frequency (like Bag-of-Words), stopwords will dominate the data, making it harder for the model to identify the rarer, more important words.Increased computation: They unnecessarily increase the size of the dataset, making training the model slower and requiring more memory.


3. Provide two real-world scenarios where removing stopwords can be harmful.

While removing stopwords is standard, it can sometimes destroy the meaning of a sentence. Two scenarios where it is harmful are:
Sentiment Analysis: Many default stopword lists include words like "not", "very", or "too". If you remove "not" from the sentence "I am not happy", it becomes "I am happy", which completely reverses the sentiment from negative to positive.
Machine Translation & Text Generation: If you are building an AI that translates languages (like Google Translate) or generates text (like ChatGPT), the model needs stopwords to understand grammar, syntax, and context so it can output human-readable, grammatically correct sentences.

4. Explain the difference between stemming and lemmatization.

Both techniques are used to reduce words to their base form, but they do it very differently:
Stemming is a fast, crude, rule-based process that simply chops the ends off of words to find the root. Because it just cuts off letters, it often results in fake words.
Example: "Running" becomes "run", but "better" might just become "bett", and "university" becomes "univers".
Lemmatization is a smarter, more advanced process that uses a dictionary and morphological analysis to find the actual linguistic root of the word (called the lemma). It requires knowing the context and part of speech.
Example: "Running" becomes "run", "better" becomes "good", and "mice" becomes "mouse".  

# Task 2: Build Advanced Preprocessing Function.

```python
import re

def preprocess_text(text):
    # Error Handling: Check if string is empty or not a string
    if not isinstance(text, str) or not text.strip():
        return [], ""

    # 1. Convert text to lowercase
    text = text.lower()

    # 2. Remove URLs and email-like patterns
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)

    # 3. Remove numbers
    text = re.sub(r'\d+', '', text)

    # 4. Handle repeated characters (Reduces 3 or more consecutive identical characters to 1)
    # Example: "soooo goooood!!!" -> "so good!!!"
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Clean up punctuation and emojis so we get clean words
    text = re.sub(r'[^\w\s]', '', text)

    # 5. Tokenization (This automatically removes extra spaces)
    raw_tokens = text.split()

    # 6. Remove short tokens (length <= 2) EXCEPT "no" and "not"
    exceptions = {"no", "not"}
    clean_tokens = [word for word in raw_tokens if len(word) > 2 or word in exceptions]

    # 7. Reconstruct the clean sentence
    clean_sentence = " ".join(clean_tokens)

    return clean_tokens, clean_sentence

print("Task 2 function successfully loaded!")
```

# Task 3: Stress Testing.

```python
# Task 3: Stress Testing

# Here are the 10 diverse sentences provided in the assignment
sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product ",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

print("--- TASK 3: STRESS TESTING RESULTS ---\n")

# We use a loop to pass each sentence one by one into our function
for sentence in sample_inputs:
    # Run our function from Task 2
    tokens, clean_sentence = preprocess_text(sentence)
    
    # Print the required output format
    print(f"Original Text   : {sentence}")
    print(f"Cleaned Tokens  : {tokens}")
    print(f"Cleaned Sentence: {clean_sentence}")
    print("-" * 50) # Adds a dividing line between sentences
```

# Task 4: Token Analytics

```python
# Task 4: Token Analytics

print("--- TASK 4: TOKEN ANALYTICS ---\n")

# Loop through the sentences again to calculate the stats
for sentence in sample_inputs:
    tokens, clean_sentence = preprocess_text(sentence)
    
    # 1. Total number of tokens
    total_tokens = len(tokens)
    
    # 2. Number of unique tokens (using 'set' removes duplicates)
    unique_tokens = len(set(tokens))
    
    # 3. Average token length
    if total_tokens > 0:
        total_characters = sum(len(word) for word in tokens)
        avg_length = total_characters / total_tokens
    else:
        avg_length = 0.0 # Just in case a sentence becomes completely empty!
        
    print(f"Original Text : {sentence}")
    print(f"Total Tokens  : {total_tokens}")
    print(f"Unique Tokens : {unique_tokens}")
    print(f"Avg Token Len : {avg_length:.2f}") # The .2f rounds it to 2 decimal places
    print("-" * 50)
```

# Task 5: Frequency Analysis

```python
# Task 5: Frequency Analysis
from collections import Counter

print("--- TASK 5: FREQUENCY ANALYSIS ---\n")

# 1. Create an empty list to hold every single token
all_tokens = []

# 2. Loop through our sentences, process them, and add the tokens to our giant list
for sentence in sample_inputs:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)

# 3. Use Counter to count how many times each word appears
token_counts = Counter(all_tokens)

# 4. Identify the Top 10 most frequent words
print("Top 10 Most Frequent Words:")
for word, count in token_counts.most_common(10):
    print(f"  - '{word}': {count} times")

print("\nTop 5 Least Frequent Words:")
# 5. To get the least frequent, we get all counts, reverse the list [::-1], and take the first 5 [:5]
least_frequent = token_counts.most_common()[::-1][:5]
for word, count in least_frequent:
    print(f"  - '{word}': {count} times")
```

# Task 6: Build Full Pipeline

```python
# Task 6: Build Full Pipeline

def full_pipeline(text_list):
    """
    Takes a list of raw text strings, processes them, and returns a dictionary 
    with the combined tokens and clean sentences.
    """
    # Create the empty dictionary with our expected output format
    pipeline_results = {
        "tokens": [],
        "clean_sentences": []
    }
    
    # Loop through the provided list
    for text in text_list:
        tokens, clean_sentence = preprocess_text(text)
        
        # Add the results to our dictionary
        pipeline_results["tokens"].append(tokens)
        pipeline_results["clean_sentences"].append(clean_sentence)
        
    return pipeline_results

print("Task 6: full_pipeline function successfully built!")
```

# Task 7: Error Handling

```python
# Task 7: Error Handling

print("--- TASK 7: ERROR HANDLING TESTS ---\n")

# These are the exact edge cases requested in the assignment
edge_cases = [
    "   ",              # Empty string / spaces only
    "😍😍😍",          # Only emojis
    "12345 67890"       # Only numbers
]

# Run our edge cases through the full pipeline from Task 6
error_handling_results = full_pipeline(edge_cases)

# Print the results to prove the pipeline didn't crash and handled them safely
for i, original in enumerate(edge_cases):
    print(f"Original Text   : '{original}'")
    print(f"Cleaned Tokens  : {error_handling_results['tokens'][i]}")
    print(f"Cleaned Sentence: '{error_handling_results['clean_sentences'][i]}'")
    print("-" * 50)
```
